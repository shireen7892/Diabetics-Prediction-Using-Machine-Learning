```python
#SPLITTING X AND Y
X=data.drop('Outcome',axis=1)#target is dependant variable
```

```python
y=data['Outcome']
X
```

```python
y
```

```python
#TRAINING
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred1=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
#randomforest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred2=rf.predict(X_test)
accuracy_score(y_test,y_pred2)
```

```python
#A confusion matrix is a table that is used to define the performance of a classification algorithm
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
grb=GradientBoostingClassifier()
grb.fit(X_train,y_train)
y_pred3=grb.predict(X_test)
accuracy_score(y_test,y_pred3)
cm = confusion_matrix(y_test, y_pred3)
cm
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
y_pred4=log.predict(X_test)
accuracy_score(y_test,y_pred4)
```

```python
from sklearn import svm
svm=svm.SVC()
svm.fit(X_train,y_train)
y_pred5=svm.predict(X_test)
accuracy_score(y_test,y_pred5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred6=knn.predict(X_test)
accuracy_score(y_test,y_pred6)
```

```python
#using find better values of n_neighbors
score=[]
for k in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred6=knn.predict(X_test)
    score.append(accuracy_score(y_test,y_pred6))
```

```python
score
```

```python
#to find our neighbouring point
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
y_pred6=knn.predict(X_test)
accuracy_score(y_test,y_pred6)
```

```python
final_data=pd.DataFrame({
    'Models':['DT','RF','GB','LR','SVM','KNN'],
    'ACC':[accuracy_score(y_test,y_pred1),
          accuracy_score(y_test,y_pred2),
          accuracy_score(y_test,y_pred3),
          accuracy_score(y_test,y_pred4),
          accuracy_score(y_test,y_pred5),
          accuracy_score(y_test,y_pred6)]
}
)
```
